```python
aff = affidavit_raw.copy()

# 1. Remove auto-index (Unnamed) columns
aff = aff.loc[:, ~aff.columns.str.contains('^Unnamed')]

# 2. Lowercase + underscore column names
aff.columns = aff.columns.str.strip().str.lower().str.replace(' ', '_')

# 3. Drop duplicates
aff = aff.drop_duplicates()

# 4. Strip and lowercase all text fields
aff['state']   = aff['state'].str.strip().str.lower()
aff['pc']      = aff['pc'].str.strip().str.lower()
aff['name']    = aff['name'].str.strip().str.lower()
aff['party']   = aff['party'].str.strip().str.lower()
aff['st_code'] = aff['st_code'].str.strip().str.upper()   # keep as S01, U01

# 5. Coerce numeric columns
aff['pc_code'] = pd.to_numeric(aff['pc_code'], errors='coerce')
aff['no']      = pd.to_numeric(aff['no'],      errors='coerce')
aff = aff.dropna(subset=['pc_code','no'])

print(f"Shape after cleaning : {aff.shape}")
print(f"Nulls                : {aff.isnull().sum().sum()}")
display(aff.head(4))

```

```python
cand = candidate_raw.copy()

# 1. Remove unnamed columns
cand = cand.loc[:, ~cand.columns.str.contains('^Unnamed')]

# 2. Standardise column names
cand.columns = cand.columns.str.strip().str.lower().str.replace(' ', '_')

# 3. Drop duplicates
cand = cand.drop_duplicates()

# 4. Lowercase text fields
cand['candidate']    = cand['candidate'].str.strip().str.lower()
cand['constituency'] = cand['constituency'].str.strip().str.lower()
cand['party']        = cand['party'].str.strip().str.lower()
cand['education']    = cand['education'].str.strip().str.lower()

# 5. Coerce numeric columns
cand['total_assets']      = pd.to_numeric(cand['total_assets'],      errors='coerce').fillna(0)
cand['total_liabilities'] = pd.to_numeric(cand['total_liabilities'], errors='coerce').fillna(0)
cand['criminal_cases']    = pd.to_numeric(cand['criminal_cases'],    errors='coerce').fillna(0)
cand['year']              = pd.to_numeric(cand['year'],              errors='coerce').astype('Int64')

# 6. Derived column: net worth
cand['net_worth'] = cand['total_assets'] - cand['total_liabilities']

# 7. Ordinal encoding for education
edu_map = {
    'illiterate': 0, 'literate': 1, '5th pass': 2, '8th pass': 3,
    '10th pass': 4,  '12th pass': 5, 'graduate': 6,
    'graduate professional': 7, 'post graduate': 8, 'doctorate': 9,
    'others': 5,     'not given': np.nan
}
cand['edu_level'] = cand['education'].map(edu_map)
cand['edu_level'] = cand['edu_level'].fillna(cand['edu_level'].median()).astype(int)

print(f"Shape after cleaning : {cand.shape}")
print(f"Nulls                : {cand.isnull().sum().sum()}")
display(cand[['candidate','constituency','education','edu_level','net_worth','year']].head(4))
```

```python
vt = votes_raw.copy()

# 1. Remove unnamed columns
vt = vt.loc[:, ~vt.columns.str.contains('^Unnamed')]

# 2. Standardise column names (replace % with pct)
vt.columns = (vt.columns.str.strip().str.lower()
                         .str.replace(' ', '_')
                         .str.replace('%', 'pct'))

# 3. Drop duplicates
vt = vt.drop_duplicates()

# 4. Lowercase text
vt['state']   = vt['state'].str.strip().str.lower()
vt['pc_name'] = vt['pc_name'].str.strip().str.lower()

# 5. Coerce integer columns
int_cols = ['male_electors','male_voters','female_electors',
            'female_voters','total_electors','total_voters']
for c in int_cols:
    vt[c] = pd.to_numeric(vt[c], errors='coerce').fillna(0).astype(int)

# 6. Coerce float columns
float_cols = ['male_turnout','female_turnout','total_turnout']
for c in float_cols:
    if vt[c].dtype == object:
        vt[c] = vt[c].str.replace('%','').astype(float)
    vt[c] = pd.to_numeric(vt[c], errors='coerce')

# 7. Derived: gender turnout gap
vt['gender_gap'] = (vt['male_turnout'] - vt['female_turnout']).round(2)

print(f"Shape after cleaning : {vt.shape}")
print(f"Nulls                : {vt.isnull().sum().sum()}")
display(vt[['state','pc_name','male_turnout','female_turnout','total_turnout','gender_gap']].head(4))
```

```python
lok = loksabha_raw.copy()

# 1. Remove unnamed columns
lok = lok.loc[:, ~lok.columns.str.contains('^Unnamed')]

# 2. Standardise column names
lok.columns = (lok.columns.str.strip().str.lower()
                          .str.replace(' ', '_')
                          .str.replace('%', 'pct'))

# 3. Drop duplicates
lok = lok.drop_duplicates()

# 4. Lowercase text fields
lok['pc_name']        = lok['pc_name'].str.strip().str.lower()
lok['state']          = lok['state'].str.strip().str.lower()
lok['candidate_name'] = lok['candidate_name'].str.strip().str.lower()
lok['party']          = lok['party'].str.strip().str.lower()
lok['type']           = lok['type'].str.strip().str.upper()   # GEN / SC / ST

# 5. Coerce numeric columns
for c in ['electors','votes','margin']:
    lok[c] = pd.to_numeric(lok[c], errors='coerce').fillna(0).astype(int)
lok['turnout']   = pd.to_numeric(lok['turnout'],   errors='coerce')
lok['marginpct'] = pd.to_numeric(lok['marginpct'], errors='coerce')
lok['year']      = lok['year'].astype(int)

# 6. Derived: won flag — rank-1 vote-getter per constituency-year
lok['vote_rank'] = (lok.groupby(['pc_name','year'])['votes']
                       .rank(method='first', ascending=False))
lok['won'] = (lok['vote_rank'] == 1).astype(int)
lok = lok.drop(columns='vote_rank')

# 7. Drop rows with missing turnout
lok = lok.dropna(subset=['turnout'])

print(f"Shape after cleaning : {lok.shape}")
print(f"Nulls                : {lok.isnull().sum().sum()}")
display(lok[['pc_name','state','year','candidate_name','votes','margin','won']].head(4))
```

```python
# Helper functions for enriched attributes
def get_decade(y): return (y // 10) * 10

def get_era(y):
    if y <= 1971:   return 'Congress Era'
    elif y <= 1989: return 'Emergency & Coalition Era'
    elif y <= 1999: return 'Coalition Instability'
    elif y <= 2009: return 'UPA Era'
    else:           return 'NDA Era'

# Build dim_year
dim_year = (lok[['year']].drop_duplicates()
                         .sort_values('year')
                         .reset_index(drop=True))

dim_year.insert(0, 'year_id', dim_year.index + 1)   # surrogate PK
dim_year['decade'] = dim_year['year'].apply(get_decade)
dim_year['era']    = dim_year['year'].apply(get_era)

print(f"dim_year → {dim_year.shape}")
display(dim_year)
```

```python
# Base: unique pc_name + state + type from loksabha
base_con = (lok[['pc_name','state','type']]
                .drop_duplicates()
                .reset_index(drop=True))

# Enrich: add st_code from affidavit
aff_geo = (aff[['st_code','state','pc']]
               .rename(columns={'pc':'pc_name'})
               .drop_duplicates(subset=['pc_name','state']))

base_con = base_con.merge(aff_geo, on=['pc_name','state'], how='left')

# Enrich: add geographic region from state name
region_map = {
    'andhra pradesh':'South', 'telangana':'South', 'tamil nadu':'South',
    'karnataka':'South', 'kerala':'South', 'puducherry':'South',
    'andaman & nicobar islands':'South',
    'maharashtra':'West', 'gujarat':'West', 'rajasthan':'West',
    'goa':'West', 'dadra & nagar haveli':'West', 'daman & diu':'West',
    'uttar pradesh':'North', 'punjab':'North', 'haryana':'North',
    'himachal pradesh':'North', 'jammu & kashmir':'North',
    'chandigarh':'North', 'delhi':'North', 'uttarakhand':'North',
    'west bengal':'East', 'odisha':'East', 'bihar':'East',
    'jharkhand':'East', 'assam':'East', 'manipur':'East',
    'meghalaya':'East', 'mizoram':'East', 'nagaland':'East',
    'sikkim':'East', 'tripura':'East', 'arunachal pradesh':'East',
    'chhattisgarh':'Central', 'madhya pradesh':'Central',
}
base_con['region'] = base_con['state'].map(region_map).fillna('Other')
base_con = base_con.reset_index(drop=True)
base_con.insert(0, 'constituency_id', base_con.index + 1)  # surrogate PK

dim_constituency = base_con[['constituency_id','pc_name','state',
                               'st_code','type','region']]

print(f"dim_constituency → {dim_constituency.shape}")
display(dim_constituency.head(6))
```

```python
# Select and rename relevant columns
dim_candidate = cand[['candidate','constituency','party','education',
                       'edu_level','criminal_cases',
                       'total_assets','total_liabilities',
                       'net_worth','year']].copy()

dim_candidate = dim_candidate.rename(columns={
    'candidate':    'candidate_name',
    'constituency': 'pc_name'
})

# Deduplicate on natural key
dim_candidate = dim_candidate.drop_duplicates(
    subset=['candidate_name','pc_name','year']).reset_index(drop=True)

# Surrogate PK
dim_candidate.insert(0, 'candidate_id', dim_candidate.index + 1)

# Derived attribute: asset tier (for OLAP slicing)
def asset_tier(v):
    if v < 500_000:       return 'Low (<5L)'
    elif v < 5_000_000:   return 'Medium (5L–50L)'
    elif v < 50_000_000:  return 'High (50L–5Cr)'
    else:                 return 'Very High (>5Cr)'

dim_candidate['asset_tier'] = dim_candidate['total_assets'].apply(asset_tier)

print(f"dim_candidate → {dim_candidate.shape}")
display(dim_candidate[['candidate_id','candidate_name','pc_name',
                        'education','criminal_cases','asset_tier','year']].head(6))
```

```python
# Unique parties from loksabha
dim_party = (lok[['party']].drop_duplicates()
                            .sort_values('party')
                            .reset_index(drop=True))

# Surrogate PK
dim_party.insert(0, 'party_id', dim_party.index + 1)

# Party type classification
national_parties = {'inc','bjp','cpi','cpim','ncp','bsp','sp','jd','jdu','rjd'}
def party_type(p):
    return 'National' if str(p).lower() in national_parties else (
           'Independent' if str(p).lower() == 'ind' else 'Regional')

dim_party['party_type'] = dim_party['party'].apply(party_type)

# Alliance mapping
alliance_map = {
    'inc':'UPA', 'bjp':'NDA', 'ncp':'UPA', 'rjd':'UPA',
    'sp':'UPA',  'bsp':'BSP', 'cpim':'LF', 'cpi':'LF',
}
dim_party['alliance'] = dim_party['party'].map(alliance_map).fillna('Other/None')

print(f"dim_party → {dim_party.shape}")
display(dim_party.head(10))
```

```python
# Select all turnout columns
dim_turnout = vt[['pc_name','state','pc_code',
                   'male_electors','female_electors',
                   'male_voters','female_voters',
                   'male_turnout','female_turnout',
                   'total_electors','total_voters',
                   'total_turnout','gender_gap']].copy()

# Deduplicate on natural key
dim_turnout = dim_turnout.drop_duplicates(
    subset=['pc_name','state']).reset_index(drop=True)

# Surrogate PK
dim_turnout.insert(0, 'turnout_id', dim_turnout.index + 1)

# Derived attribute: turnout category
def turnout_cat(t):
    if t < 50:    return 'Low (<50%)'
    elif t < 65:  return 'Medium (50–65%)'
    elif t < 75:  return 'High (65–75%)'
    else:         return 'Very High (>75%)'

dim_turnout['turnout_category'] = dim_turnout['total_turnout'].apply(turnout_cat)

print(f"dim_turnout → {dim_turnout.shape}")
display(dim_turnout[['turnout_id','pc_name','state','male_turnout',
                      'female_turnout','total_turnout','turnout_category']].head(6))
```

```python
dim_summary = pd.DataFrame({
    'Table'        : ['dim_year','dim_constituency','dim_candidate',
                      'dim_party','dim_turnout'],
    'Primary Key'  : ['year_id','constituency_id','candidate_id',
                      'party_id','turnout_id'],
    'Source'       : ['loksabha','loksabha + affidavit','candidate',
                      'loksabha','votes_turnout'],
    'Rows'         : [len(dim_year), len(dim_constituency), len(dim_candidate),
                      len(dim_party), len(dim_turnout)],
    'Columns'      : [dim_year.shape[1], dim_constituency.shape[1],
                      dim_candidate.shape[1], dim_party.shape[1], dim_turnout.shape[1]]
})
display(dim_summary)
print("\n✅ All 5 dimension tables ready.")
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
ax.set_xlim(0, 15)
ax.set_ylim(0, 9)
ax.axis('off')
fig.patch.set_facecolor('#0d1117')
ax.set_facecolor('#0d1117')

def draw_box(ax, x, y, w, h, title, rows, t_color, b_color):
    rect = mpatches.FancyBboxPatch(
        (x, y), w, h, boxstyle="round,pad=0.1",
        linewidth=2, edgecolor=t_color, facecolor=b_color)
    ax.add_patch(rect)
    ax.text(x+w/2, y+h-0.22, title, ha='center', va='top',
            fontsize=8.5, fontweight='bold', color=t_color)
    ax.plot([x+0.1, x+w-0.1], [y+h-0.38, y+h-0.38],
            color=t_color, lw=0.8, alpha=0.7)
    step = (h - 0.48) / max(len(rows), 1)
    for i, r in enumerate(rows):
        ax.text(x+0.18, y+h-0.54-i*step, r, ha='left', va='top',
                fontsize=6.8, color='#e6edf3', fontfamily='monospace')

def arrow(ax, x1, y1, x2, y2, c='#58a6ff'):
    ax.annotate('', xy=(x2,y2), xytext=(x1,y1),
        arrowprops=dict(arrowstyle='->', color=c, lw=1.8))

# ── FACT TABLE ─────────────────────────────────
draw_box(ax, 5.2, 3.0, 4.2, 3.2,
    '★  fact_election_results',
    ['result_id         PK',
     '─────────────────────',
     'constituency_id   FK ──→ dim_constituency',
     'year_id           FK ──→ dim_year',
     'party_id          FK ──→ dim_party',
     'candidate_id      FK ──→ dim_candidate',
     'turnout_id        FK ──→ dim_turnout',
     '─────────────────────',
     'votes        (measure)',
     'electors     (measure)',
     'margin       (measure)',
     'margin_pct   (measure)',
     'turnout      (measure)',
     'won          (measure)'],
    '#f0c040', '#1c2333')

# ── DIM YEAR ────────────────────────────────────
draw_box(ax, 5.8, 7.0, 3.1, 1.7,
    'dim_year',
    ['year_id  PK', 'year', 'decade', 'era'],
    '#58a6ff', '#161b22')
arrow(ax, 7.35, 7.0, 7.35, 6.2)

# ── DIM CONSTITUENCY ────────────────────────────
draw_box(ax, 0.2, 3.5, 3.5, 2.4,
    'dim_constituency',
    ['constituency_id  PK', 'pc_name', 'state',
     'st_code', 'type', 'region'],
    '#3fb950', '#161b22')
arrow(ax, 3.7, 4.7, 5.2, 4.7)

# ── DIM CANDIDATE ───────────────────────────────
draw_box(ax, 11.2, 3.5, 3.5, 2.8,
    'dim_candidate',
    ['candidate_id  PK', 'candidate_name', 'pc_name',
     'education', 'edu_level',
     'criminal_cases', 'net_worth', 'asset_tier'],
    '#da70d6', '#161b22')
arrow(ax, 11.2, 4.9, 9.4, 4.9)

# ── DIM PARTY ───────────────────────────────────
draw_box(ax, 0.2, 0.4, 3.5, 2.2,
    'dim_party',
    ['party_id   PK', 'party',
     'party_type', 'alliance'],
    '#ff7b72', '#161b22')
arrow(ax, 3.7, 1.8, 5.5, 3.2)

# ── DIM TURNOUT ─────────────────────────────────
draw_box(ax, 11.2, 0.4, 3.5, 2.8,
    'dim_turnout',
    ['turnout_id     PK', 'pc_name', 'state',
     'male_electors', 'female_electors',
     'male_turnout', 'female_turnout',
     'total_turnout', 'turnout_category'],
    '#ffa657', '#161b22')
arrow(ax, 11.2, 1.8, 9.4, 3.2)

ax.text(7.5, 8.85,
        'LOK SABHA ELECTION — STAR SCHEMA',
        ha='center', va='top', fontsize=15,
        fontweight='bold', color='#f0c040')
ax.text(7.5, 8.42,
        '1 Fact Table  ·  5 Dimension Tables  ·  Foreign Key links shown with arrows',
        ha='center', va='top', fontsize=9, color='#8b949e')

plt.tight_layout()
plt.savefig(f"{OUTPUT}/star_schema_diagram.png",
            bbox_inches='tight', dpi=160, facecolor='#0d1117')
plt.show()
print("✅ Diagram saved.")
```

```python
df = pd.read_sql("""
    SELECT
        f.result_id,
        f.votes,      f.electors,  f.margin,
        f.marginpct,  f.turnout,   f.won,
        c.pc_name,    c.state,     c.type,
        c.region,
        p.party,      p.party_type,p.alliance,
        y.year,       y.decade,    y.era,
        d.education,  d.edu_level,
        d.criminal_cases,          d.net_worth,
        d.asset_tier,
        t.male_turnout, t.female_turnout, t.gender_gap,
        t.turnout_category
    FROM   fact_election_results f
    JOIN   dim_constituency c  ON f.constituency_id = c.constituency_id
    JOIN   dim_party        p  ON f.party_id        = p.party_id
    JOIN   dim_year         y  ON f.year_id         = y.year_id
    LEFT JOIN dim_candidate d  ON f.candidate_id    = d.candidate_id
    LEFT JOIN dim_turnout   t  ON f.turnout_id      = t.turnout_id
    WHERE  c.region IS NOT NULL
      AND  f.turnout IS NOT NULL
""", conn).dropna(subset=["type","region","party_type","decade","turnout"])

df["net_worth"]      = df["net_worth"].fillna(0)
df["criminal_cases"] = df["criminal_cases"].fillna(0)
df["edu_level"]      = df["edu_level"].fillna(df["edu_level"].median() if "edu_level" in df else 6)
df["gender_gap"]     = df["gender_gap"].fillna(0)

# ── Encode categorical columns on full df (needed for regression) ─
le_df = {}
for col in ["type","region","party_type","alliance","era"]:
    le_df[col] = LabelEncoder()
    df[col+"_enc"] = le_df[col].fit_transform(df[col].fillna("Unknown"))

print(f"✅ Dataset loaded : {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"   Year range     : {int(df['year'].min())} - {int(df['year'].max())}")
print(f"   Regions        : {sorted(df['region'].dropna().unique().tolist())}")
display(df.head(4))
```

```python
# ── Standard library ────────────────────────────────────────────
import warnings
warnings.filterwarnings("ignore")
import os, sqlite3, itertools
from collections import defaultdict

import pandas as pd
import numpy  as np

# ── Visualisation ────────────────────────────────────────────────
import matplotlib.pyplot  as plt
import matplotlib.patches as mpatches
import seaborn as sns

# ── Machine Learning ─────────────────────────────────────────────
from sklearn.model_selection  import (train_test_split, cross_val_score,
                                      StratifiedKFold, GridSearchCV)
from sklearn.preprocessing    import LabelEncoder, StandardScaler, PolynomialFeatures
from sklearn.tree             import DecisionTreeClassifier, export_text
from sklearn.ensemble         import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes      import GaussianNB
from sklearn.cluster          import KMeans, AgglomerativeClustering
from sklearn.decomposition    import PCA
from sklearn.linear_model     import LinearRegression, Ridge, Lasso
from sklearn.pipeline         import Pipeline
from sklearn.metrics          import (accuracy_score, classification_report,
                                      confusion_matrix, ConfusionMatrixDisplay,
                                      roc_curve, auc, silhouette_score,
                                      mean_squared_error, r2_score,
                                      adjusted_rand_score)

# ── Config ───────────────────────────────────────────────────────
SEED   = 42
OUTPUT = "star_schema_output"     # CSVs from Part 1
OUT    = "model_training_output"  # plots from Part 5
DB     = "loksabha_star_schema.db"

os.makedirs(OUTPUT, exist_ok=True)
os.makedirs(OUT,    exist_ok=True)

plt.rcParams.update({"figure.dpi": 130, "font.size": 9,
                     "axes.titlesize": 11, "axes.labelsize": 9})

print("✅ All libraries loaded.")
print(f"📁 Star schema output : {OUTPUT}/")
print(f"📁 Model output       : {OUT}/")
print(f"🗄️  SQLite DB          : {DB}")
```

```python
winners = df[df["won"] == 1].copy()
SAFE_THRESHOLD = 15.0
winners["constituency_type"] = np.where(
    winners["marginpct"] >= SAFE_THRESHOLD, "Safe", "Swing")
dist = winners["constituency_type"].value_counts()
print(f"Safe  : {dist.get('Safe',0):>5} ({100*dist.get('Safe',0)/len(winners):.1f}%)")
print(f"Swing : {dist.get('Swing',0):>5} ({100*dist.get('Swing',0)/len(winners):.1f}%)")
print(f"Total : {len(winners):>5}")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].bar(["Safe","Swing"], [dist.get("Safe",0), dist.get("Swing",0)],
            color=["#3fb950","#ff7b72"], edgecolor="white", width=0.5)
axes[0].set_title("Constituency Type Distribution\n(Margin >=15% = Safe, <15% = Swing)")
axes[0].set_ylabel("Count")
for i, (label, val) in enumerate([("Safe",dist.get("Safe",0)),("Swing",dist.get("Swing",0))]):
    axes[0].text(i, val+20, f"{val:,}\n({100*val/len(winners):.1f}%)", ha="center", fontsize=9)
safe_m  = winners[winners["constituency_type"]=="Safe"]["marginpct"]
swing_m = winners[winners["constituency_type"]=="Swing"]["marginpct"]
axes[1].hist(swing_m, bins=30, color="#ff7b72", alpha=0.75, label="Swing (<15%)")
axes[1].hist(safe_m,  bins=30, color="#3fb950", alpha=0.75, label="Safe  (>=15%)")
axes[1].axvline(SAFE_THRESHOLD, color="black", lw=2, linestyle="--")
axes[1].set_title("Margin % Distribution — Safe vs Swing")
axes[1].set_xlabel("Winning Margin (%)"); axes[1].legend(fontsize=8)
plt.suptitle("Safe vs Swing Constituency Definition", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.savefig(f"{OUT}/A1_safe_swing_distribution.png", dpi=130)
plt.show()
print("\n✅ Target label created.")
```

```python
le = {}
for col in ["type","region","party_type","alliance","era"]:
    le[col] = LabelEncoder()
    winners[col+"_enc"] = le[col].fit_transform(winners[col].fillna("Unknown"))
winners["target"] = (winners["constituency_type"] == "Safe").astype(int)
FEATURES = ["votes","electors","turnout","type_enc","region_enc",
            "party_type_enc","alliance_enc","era_enc","decade"]
for opt in ["edu_level","criminal_cases","net_worth","gender_gap"]:
    if winners[opt].notna().sum() > 100:
        FEATURES.append(opt)
X = winners[FEATURES].fillna(0)
y = winners["target"]
print(f"Feature set ({len(FEATURES)} features):")
for i, f in enumerate(FEATURES, 1): print(f"  {i:>2}. {f}")
print(f"\nX shape : {X.shape}")
print(f"y balance: Safe={y.sum():,}  Swing={(y==0).sum():,}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=SEED, stratify=y)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)
print(f"Training set : {len(X_train):>6,}  (Safe={y_train.sum():,}  Swing={(y_train==0).sum():,})")
print(f"Test set     : {len(X_test):>6,}  (Safe={y_test.sum():,}   Swing={(y_test==0).sum():,})")
```

```python
dt = DecisionTreeClassifier(max_depth=8, min_samples_split=10,
                             min_samples_leaf=5, criterion="gini", random_state=SEED)
dt.fit(X_train_s, y_train)
y_pred_dt = dt.predict(X_test_s)
acc_dt    = accuracy_score(y_test, y_pred_dt)
print("=" * 55)
print("  MODEL 1 — Decision Tree Classifier")
print("=" * 55)
print(f"  Accuracy : {acc_dt:.4f} ({acc_dt*100:.2f}%)")
print()
print(classification_report(y_test, y_pred_dt, target_names=["Swing","Safe"]))
print("\nDecision Tree Rules (first 3 levels):")
print(export_text(dt, feature_names=FEATURES, max_depth=3))
```

```python
rf = RandomForestClassifier(n_estimators=200, max_depth=12, random_state=SEED, n_jobs=-1)
rf.fit(X_train_s, y_train)
y_pred_rf = rf.predict(X_test_s)
y_prob_rf = rf.predict_proba(X_test_s)[:, 1]
acc_rf    = accuracy_score(y_test, y_pred_rf)
print("=" * 55)
print("  MODEL 2 — Random Forest Classifier")
print("=" * 55)
print(f"  Accuracy : {acc_rf:.4f} ({acc_rf*100:.2f}%)")
print()
print(classification_report(y_test, y_pred_rf, target_names=["Swing","Safe"]))
fi = pd.Series(rf.feature_importances_, index=FEATURES).sort_values(ascending=False)
print("\nTop 5 Features:")
print(fi.head(5).round(4).to_string())
```

```python
gb = GradientBoostingClassifier(n_estimators=150, learning_rate=0.1,
                                   max_depth=5, random_state=SEED)
gb.fit(X_train_s, y_train)
y_pred_gb = gb.predict(X_test_s)
y_prob_gb = gb.predict_proba(X_test_s)[:, 1]
acc_gb    = accuracy_score(y_test, y_pred_gb)
print("=" * 55)
print("  MODEL 3 — Gradient Boosting Classifier")
print("=" * 55)
print(f"  Accuracy : {acc_gb:.4f} ({acc_gb*100:.2f}%)")
print()
print(classification_report(y_test, y_pred_gb, target_names=["Swing","Safe"]))
```
